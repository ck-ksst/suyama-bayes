# 2.2.1 ベルヌーイ分布

```python
def Bern(x, mu):
    """
    ベルヌーイ分布（定義通りです）
    x: 0 or 1
    mu: 1が出る確率
    """
    return mu ** x * (1 - mu) ** (1 - x)
```

```python
def ln_Bern(x, mu):
    return x * np.log(mu) + (1 - x) * np.log(1 - mu)
```

```python
mu = 0.7 # 1が出る確率
x = np.array([0., 1.]) # 裏か表か．
prob = np.array([1 - mu, mu]) # 裏が出る確率と表が出る確率
```

```python
plt.bar(x, prob)
plt.xlabel('x')
plt.ylabel('確率')
plt.title('ベルヌーイ分布')
plt.xticks(ticks=[0, 1])
plt.legend()
```

# 2.2.2 二項分布

```python
def Bin(m, M, mu):
    """
    m: 表(1)が出る回数
    M: 試行回数
    mu: 表(1)が出る確率
    """
    return comb(M, m, exact=True) * (mu ** m) * ((1 - mu) ** (M - m))
```

```python
M = 10
mu = 0.5
m = np.arange(0, M, 1)
p = np.array([Bin(m, M, mu) for m in range(M)])
```

# 2.2.3 カテゴリ分布

```python
def Cat(s, pi):
    prob = 1
    for k in range(len(pi)):
        prob *= pi[k] ** s[k]
    return prob
```

# 2.2.4 多項分布

```python
def Mult(m, pi):
    ''' 
    m:  それぞれの出現回数のリスト
    pi: それぞれの出現確率
    M:  試行回数
    '''
    M = np.sum(m)
    ans = 1
    for k in range(len(m)):
        ans *= (pi[k] ** m[k]) / factorial(m[k])
    ans *= factorial(M)
    return ans
```

# 2.2.5 ポアソン分布

```python
def Poi(x, lmd):    
    return np.array([(lmd ** x[i]) * np.exp(-lmd) / factorial(x[i]) for i in range(len(x))])
```

```python
x = np.arange(0, 13, 1)
```

```python
lmd = 0.5
p = Poi(x, lmd)
plt.bar(x, p)
```

```python
lmd = 4.0
p = Poi(x, lmd)
plt.bar(x, p)
```
